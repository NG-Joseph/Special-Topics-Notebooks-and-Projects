# EDA - Exploratory Data Analysis STEP 3

```python
#Under EDA the first step is to search for missing values
#to do this we use:
train_data.isnull().sum() 
#we could have stopped after isnull but we addws thw sum function yo show the number of missing values in each row. Without the sum function it would have just gone through and come out with yes or no responses.
```

```python
#Thankfully we have no missing values so we do not need to drop any column yet.
#Now we have to go through our data and check for values that are not numeric. 
#ML only understands numeric values so we have to do something to change values that are not numeric.
#First let's examine a sample from our data set.
train_data.head()
```

```python
#From the above, we can see that surname,geography and gender have values that are not numeric.
#Now we move to the next step under EDA which is data preprocessing. 
#To achieve this we will be using a library called sklearn
#from that library, we will use a tool called labelencoder
#There are two ways to encode non-numeric variables but we will be using the label encoding method
#We're going to import our labelencoder function from sklearn
from sklearn.preprocessing import LabelEncoder
```

```python
#For easy referencing, we will assign labelencoder to a variable
le = LabelEncoder()
```

```python
train_data
```

```python
train_data["Surname"] = le.fit_transform(train_data["Surname"])
train_data["Gender"] = le.fit_transform(train_data["Gender"])
train_data["Geography"] = le.fit_transform(train_data["Geography"])
```

```python
#Let's do the same for the test data
```
