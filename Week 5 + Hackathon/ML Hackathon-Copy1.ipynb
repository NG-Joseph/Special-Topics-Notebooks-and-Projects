```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('train_data.csv')
test_data = pd.read_csv('test_data.csv')
```

```python
train_data
```

```python
test_data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
train_data[["Geography", "Gender"]] = train_data[["Geography", "Gender"]].apply(le.fit_transform)
test_data[["Geography", "Gender"]] = test_data[["Geography", "Gender"]].apply(le.fit_transform)
train_data
```

```python
fig, ax = plt.subplots(figsize=(16,8))
ax.scatter(train_data['Balance'], train_data['Balance'])
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

```python
x_train = train_data[["CreditScore", "Age", "Gender", "Geography", "EstimatedSalary", "NumOfProducts", "IsActiveMember",  "Balance"]].values
y_train = train_data['Exited'].values
x_test = test_data[["CreditScore", "Age", "Gender", "Geography", "EstimatedSalary", "NumOfProducts", "IsActiveMember", "Balance"]].values
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
#classifier = RandomForestClassifier()
#classifier.fit(x_train, y_train)
classifier2 = SVC(kernel='linear')
```

```python
classifier2.fit(x_train, y_train)
```

```python
y_prediction = classifier2.predict(x_test)
```

```python
result_dict = {'CustomerId': test_data["CustomerId"], 'Exited':y_prediction} 
```

```python
result = pd.DataFrame(result_dict)
```
