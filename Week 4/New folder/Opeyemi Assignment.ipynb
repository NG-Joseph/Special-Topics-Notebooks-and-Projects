```python
import pandas as pd
```

```python
dataset= pd.read_csv('NYC_Property_Data.csv')
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(how='any', inplace=True)
```

```python
dataset.isnull().sum()
```

```python
features = dataset.drop("SALE PRICE", axis=1)
features
```

```python
target = dataset["SALE PRICE"]
target
```

```python
features = dataset.drop(["SALE DATE","BLOCK", "LOT", "ZIP CODE", "Unnamed: 0", "Unnamed: 0.1" ], axis=1)
features
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
features["ADDRESS"] = le.fit_transform(features["ADDRESS"])
```

```python
features
```

```python
features["BUILDING CLASS CATEGORY"] = le.fit_transform(features["BUILDING CLASS CATEGORY"])
```

```python
features
```

```python
features["NEIGHBORHOOD"] = le.fit_transform(features["NEIGHBORHOOD"])
```

```python
features
```

```python
features["TAX CLASS AT PRESENT"] = le.fit_transform(features["TAX CLASS AT PRESENT"])
```

```python
features
```

```python
features["BUILDING CLASS AT PRESENT"] = le.fit_transform(features["BUILDING CLASS AT PRESENT"])
```

```python
features
```

```python
features["BUILDING CLASS AT TIME OF SALE"] = le.fit_transform(features["BUILDING CLASS AT TIME OF SALE"])
```

```python
features
```

```python
from sklearn.model_selection import train_test_split
```

```python
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.3)
```

```python
features_train
```

```python
features_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression() #lm = linear model
```

```python
model = lm.fit(X=features_train, y=target_train)
```

```python
prediction = model.predict(features_test)
```

```python
result = pd.DataFrame({"Actual Value":target_test, "Predicted Value": prediction, "Error Margin": abs(target_test - prediction)})
```
