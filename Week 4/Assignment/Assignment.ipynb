```python
features = data.drop("SALE PRICE", axis=1) # Separate features from target
```

```python
features = features[["BOROUGH", "NEIGHBORHOOD", "BUILDING CLASS CATEGORY", "BUILDING CLASS AT PRESENT", "TAX CLASS AT PRESENT", "RESIDENTIAL UNITS", "COMMERCIAL UNITS", "TOTAL UNITS", "LAND SQUARE FEET", "GROSS SQUARE FEET", "YEAR BUILT", "TAX CLASS AT TIME OF SALE", "BUILDING CLASS AT TIME OF SALE"]]
```

```python
features
```

```python
target = data["SALE PRICE"] # Assign sale price to target 
```

## Next Step: Convert categorical data into labels (Label Encoder)

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
type(data["TAX CLASS AT PRESENT"][0]) # Check type of Tax class at present
```

```python
features[["NEIGHBORHOOD",  "TAX CLASS AT PRESENT", "BUILDING CLASS CATEGORY","BUILDING CLASS AT TIME OF SALE","BUILDING CLASS AT PRESENT", "ADDRESS"]]= features[["NEIGHBORHOOD","BUILDING CLASS AT PRESENT", "BUILDING CLASS AT TIME OF SALE", "TAX CLASS AT PRESENT","BUILDING CLASS AT TIME OF SALE", "BUILDING CLASS CATEGORY"]].apply(le.fit_transform) #= le.fit_transform(features[["NEIGHBORHOOD", "TAX CLASS AT PRESENT","BUILDING CLASS CATEGORY", "ADDRESS"]])
```

```python
from sklearn.model_selection import train_test_split
```

```python
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.2)
```

```python
features_test.shape
```

```python
target_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression() #lm = linear model
```

```python
model = lm.fit(X=features_train, y=target_train)
```

```python
features_test
```

```python
prediction = model.predict(features_test)
```

```python
result = pd.DataFrame({"Actual Value":target_test, "Predicted Value": prediction, "Error Margin": abs(target_test - prediction)})
```
