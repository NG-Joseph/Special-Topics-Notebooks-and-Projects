## Which Director's movies made the most revenue in the dataset?

```python
revenue_by_director = imdb_data.groupby('Director')[['Revenue (Millions)']].sum()
```

```python
revenue_by_director.sort_values(by="Revenue (Millions)", ascending=False)
```

## J.J Abrams Films made the most revenue
