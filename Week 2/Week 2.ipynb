# Week 2: Data Science

### Definition
Data science is an interdisciplinary field that uses scientific methods, processes, algorithms and systems to extract knowledge and insights from noisy, structured and unstructured data, and apply knowledge and actionable insights from data across a broad range of application domains


### Types of Data
- Qualitative Data
- Quantitative Data

### Quantitative Data
This consists of:
- **Categorical Data**: Consists of Categorical Nominal (Eye color), Categorical Ordinal (age group) and Categorical Binary (Male or Female)
- **Discrete**: Number of students, age. Can be likened to integer values
- **Continuous**: Takes in a range of values (you can liken it to float values)

### Qualitative Data
This provides information about the quality of a good or service.
- Customer Feedback
- Group Discussion
- Answers to interviews


### Shapes of Data
Data comes in three major shapes 
- Structured Data: These are quantitative data.
- Unstructured Data: For example, emails.
- Semi-Structured Data: Comma seperated values for example. An inherent structure but poorly represented format usually separated by a delimiter such as comma ','.

### Situating Data Science
Data Science consists of 
- Computing 
- Domain Knowledge 
- Maths and Statistics

### Stages of Data Science Methodology
- From problem to approach
- From Requirements to collectoin
- From Understanding to Preparation
- From Modeling to Evaluation
- From Deployment to Feedback

### What Data Science is About
- Asking the correct questions and analyzing the raw data
- Modelling the data using various complex and efficient algorithms
- Visualizing data to get a better perspective
- Analyzing data to make better business decisions

### Questions for Data Science Projects
- What problem are you trying to solve?
- How can you use data to answer the questions?
- What data do you need to answer the question?
- Where is the data coming from and how will you get it?
- What additional work is required to manipulate and work with data?
- In what way can the data be visualized to get to the answer that is required?
- Can you put the model into practice?
- Can you get constructive feedback into answering the question?
