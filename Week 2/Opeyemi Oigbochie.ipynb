1. Find the movies that have adventure in it and their Revenue 
2. Find the movies that Will Smith acted in and had a high rating (>=7)
3. Find the movies that start with 'S' and their minutes (< 90)

```python
import pandas as pd
```

```python
data = pd.read_csv('IMDB-Movie-Data.csv')
```

```python
data
```

```python
data.dropna(how='any', inplace=True)
```

```python
data
```

```python
q1 = data[data["Genre"].str.contains("Action")]
```

```python
q1[["Title", "Revenue (Millions)"]]
```

```python
q2 = data[data["Actors"].str.contains("Will Smith")]
```

```python
q2
```

```python
 q2[q2["Rating"] >= 7]
```

```python
q3 = data[data["Title"].str.startswith("s") | data["Title"].str.startswith("S") ]
```

```python
q3
```

```python
q3[q3["Runtime (Minutes)"] < 90]
```
