```python
import glob 

all_files = glob.glob("./*.csv")
all_files
```

```python
li = []
for data in all_files: 
    df = pd.read_csv(data)
    li.append(df)
    
    
```

```python
data = pd.concat(li,axis=0, ignore_index=True)
```

## Best time to display ads 
Ads can be used to promote sales at periods where they tend to get less. We can uncover this by finding the month with the least sales

```python
data
```

```python
data_by_month = data
```

```python
data_by_month
```

```python
data_by_month.describe()
```

```python
data_by_month = data_by_month.groupby(by='Order Month')
```

```python
data["Quantity Ordered"] = data["Quantity Ordered"].astype(int)
```

```python
quantity_bought_by_month = data.groupby('Order Month')[['Quantity Ordered']].sum()
```

```python
quantity_bought_by_month.sort_values(by="Quantity Ordered", ascending=True)
```
